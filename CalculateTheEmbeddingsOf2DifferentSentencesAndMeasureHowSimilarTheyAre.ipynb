```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
 # load Model and tokenize
modelo = AutoModel.from_pretrained("bert-base-uncased")
tokenizador = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
# Phrases to compare
frase_1 = "the cat is on the mat."
frase_2 = "Wi√±apacha"
```

```python
 # Encode phrases
inputs_1 = tokenizador(frase_1, return_tensors="pt")
inputs_2 = tokenizador(frase_2, return_tensors="pt")
```

```python
# Generate embeddings
output_1 =  modelo(**inputs_1).last_hidden_state
output_2 =  modelo(**inputs_2).last_hidden_state
```

```python
 # Get embeddings of the entire phrase (average of tokens)
embedding_1 = output_1.mean(dim=1).detach().numpy()
embedding_2 = output_2.mean(dim=1).detach().numpy()
```

```python
 # Calculate cosine similarity
similarity = cosine_similarity(embedding_1, embedding_2)
print(f"Similarity between phrases: {similarity[0][0]:.2f}")
```

```python
Similarity between phrases: 0.39
```
